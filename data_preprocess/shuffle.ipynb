```python
import tensorflow as tf
import time
from PIL import Image
from tqdm import tqdm
```

```python
shuffle_paths = {
    "train1": "/content/drive/Shareddrives/DL_Final_Project/shuffle/train1",
    "test1": "/content/drive/Shareddrives/DL_Final_Project/shuffle/test1",
    "train2": "/content/drive/Shareddrives/DL_Final_Project/shuffle/train2",
    "test2": "/content/drive/Shareddrives/DL_Final_Project/shuffle/test2",
    "train3": "/content/drive/Shareddrives/DL_Final_Project/shuffle/train3",
    "test3": "/content/drive/Shareddrives/DL_Final_Project/shuffle/test3",
}
```

```python
train_dataset = tf.data.Dataset.load(shuffle_paths["train3"])
batch_size = 256
train_dataset = train_dataset.batch(batch_size)
train_dataset = train_dataset.prefetch(tf.data.experimental.AUTOTUNE)
```

```python
print(len(train_dataset))
```

```python
count = 0
out_count = 0
for batch_images, batch_labels in train_dataset:
  if out_count == 0:
    for image, label in zip(batch_images, batch_labels):
      if count < 20:
        img = tf.keras.utils.array_to_img(image)
        print("image")
        display(img)
        print("label")
        print(tf.math.argmax(label))
        count += 1
      else:
        break
    out_count += 1
  else:
    break
```
